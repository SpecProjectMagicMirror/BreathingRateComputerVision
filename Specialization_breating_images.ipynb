```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
cap = cv2.VideoCapture('/Users/arpitsabherwal/Desktop/FLIR0004.mov')
nose_cascade = cv2.CascadeClassifier('/Users/arpitsabherwal/Downloads/MMM-nose/cascade_files/Nariz.xml')

if nose_cascade.empty():
  raise IOError('Unable to load the nose cascade classifier xml file')
```

```python
img_thermal = cv2.imread('/Users/arpitsabherwal/Desktop/thermal/FLIR0008.jpg')
```

```python
img_normal=cv2.imread('/Users/arpitsabherwal/Desktop/normal/new_normal_1.png')
```

```python
count =0
y_val=[]
i=0
while(i==0):
    gray = cv2.cvtColor(img_normal, cv2.COLOR_BGR2GRAY)
    nose_rects = nose_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in nose_rects:
        print("true")
        frame1=cv2.rectangle(img_thermal, (x,y), (x+w+30,y+h), (0,255,0), 3)
        tmp1 = cv2.cvtColor(img_thermal[y:y+h,x:x+w+30], cv2.COLOR_BGR2GRAY)
        histeq = cv2.equalizeHist(tmp1)
        plt.imshow(histeq)
        y_val.append(np.average(histeq))
        name = "frames/frame%d.jpg"%count
        cv2.imwrite(name, frame1)
        count=count+1
        i=1
        break
```
